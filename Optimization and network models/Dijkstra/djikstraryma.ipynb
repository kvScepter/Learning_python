```python
import networkx as nx 
import matplotlib.pyplot as plt
import numpy as np

A = np.loadtxt(open("AA5360_kaupungit.csv", "rb"), delimiter=",", skiprows=0)
G = nx.from_numpy_matrix(A)

#G = nx.Graph()
print('linkkien lukumäärä:', G.number_of_edges())
print('Noodien lukumäärä:', G.number_of_nodes())
mylayout = nx.circular_layout(G)
nx.draw(G, mylayout, with_labels=True)
nx.write_graphml(G, 'DJIK2.graphml')
```
